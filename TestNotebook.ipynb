# Data Augmentation

```python
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt
# Function to perform data augmentation on a single image
def augment_image(image):
    # Random rotation (between -10 and 10 degrees)
    angle = random.randint(-10, 10)
    rows, cols, _ = image.shape
    M = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
    image = cv2.warpAffine(image, M, (cols, rows))

    # Random horizontal flip
    if random.random() < 0.5:
        image = cv2.flip(image, 1)

    # Random vertical flip
    if random.random() < 0.5:
        image = cv2.flip(image, 0)

    # Random brightness adjustment (factor between 0.7 and 1.3)
    brightness_factor = random.uniform(0.7, 1.3)
    image = cv2.convertScaleAbs(image, alpha=brightness_factor, beta=0)

    return image

# Read the single image
image_path = "elephant.jpg"
image = cv2.imread(image_path)

# Perform augmentation
augmented_image = augment_image(image)

# Display the original and augmented images side by side (for visualization)
combined_image = np.hstack((image, augmented_image))
# cv2.imshow("Original vs Augmented", combined_image)
plt.imshow(cv2.cvtColor(combined_image,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import os
import random

# Function to perform data augmentation on a single image
def augment_image(image):
    # Random rotation (between -10 and 10 degrees)
    angle = random.randint(-10, 10)
    rows, cols, _ = image.shape
    M = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
    image = cv2.warpAffine(image, M, (cols, rows))

    # Random horizontal flip
    if random.random() < 0.5:
        image = cv2.flip(image, 1)

    # Random vertical flip
    if random.random() < 0.5:
        image = cv2.flip(image, 0)

    # Random brightness adjustment (factor between 0.7 and 1.3)
    brightness_factor = random.uniform(0.7, 1.3)
    image = cv2.convertScaleAbs(image, alpha=brightness_factor, beta=0)

    return image

# Read the single image
image_path = "elephant.jpg"
image = cv2.imread(image_path)

# Number of augmented images to generate
num_augmented_images = 5

# Create a directory to save the augmented images
output_dir = "dataAug"
os.makedirs(output_dir, exist_ok=True)

# Apply multiple data augmentations and save the resultant images
for i in range(num_augmented_images):
    augmented_image = augment_image(image)
    output_path = os.path.join(output_dir, f"augmented_{i+1}.jpg")
    cv2.imwrite(output_path, augmented_image)
```

```python
# CV2 data preprocessing
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image_path = "elephant.jpg"
image = cv2.imread(image_path)

# Resize the image to a specific width and height
target_width = 300
target_height = 200
resized_image = cv2.resize(image, (target_width, target_height))

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

# Apply edge detection using Canny
canny_edges = cv2.Canny(gray_image, threshold1=50, threshold2=150)

# Convert the image to HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Extract the individual channels (Hue, Saturation, Value)
h, s, v = cv2.split(hsv_image)

# Perform histogram equalization on the Value channel
equalized_v = cv2.equalizeHist(v)

# Merge the equalized Value channel back with the original Hue and Saturation channels
equalized_hsv = cv2.merge([h, s, equalized_v])

# Convert the HSV image back to BGR color space
equalized_bgr = cv2.cvtColor(equalized_hsv, cv2.COLOR_HSV2BGR)

# Save the preprocessed images
cv2.imwrite("path/to/save/resized_image.jpg", resized_image)
cv2.imwrite("path/to/save/gray_image.jpg", gray_image)
cv2.imwrite("path/to/save/blurred_image.jpg", blurred_image)
cv2.imwrite("path/to/save/canny_edges.jpg", canny_edges)
cv2.imwrite("path/to/save/equalized_bgr.jpg", equalized_bgr)

# Display all the preprocessed images in subplots
titles = ["Original", "Resized", "Grayscale", "Blurred", "Canny Edges", "Equalized BGR"]
images = [image, resized_image, gray_image, blurred_image, canny_edges, equalized_bgr]

num_images = len(images)
rows = 2
cols = (num_images + 1) // rows

plt.figure(figsize=(12, 8))

for i in range(num_images):
    plt.subplot(rows, cols, i+1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

# To capture video feed from the live stream

```python
"""
To capture frames from a video stream, you can use OpenCV to read frames from a video source 
(e.g., a camera feed or a video file) 
and then save selected frames into a folder. Below is a Python code snippet that captures frames 
from a video stream and saves every 50th frame into a folder:

"""

import cv2
import os

# Create a VideoCapture object with the correct camera index or video file path
cap = cv2.VideoCapture(0)  # Use 0 for the default camera (HDMI input) or specify the video file path

# Check if the video source is opened correctly
if not cap.isOpened():
    raise IOError("Error: Cannot open the video source.")

# Create a folder to store the captured frames (if it doesn't exist)
output_folder = "captured_frames"
os.makedirs(output_folder, exist_ok=True)

# Frame count and frame interval for capturing (e.g., save every 50th frame)
frame_count = 0
frame_interval = 50

# Loop to continuously capture frames
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Check if the frame is read correctly
    if not ret:
        break

    # Increment frame count
    frame_count += 1

    # Save the frame if it's the 50th frame or a multiple of the interval
    if frame_count % frame_interval == 0:
        frame_filename = os.path.join(output_folder, f"frame_{frame_count:04d}.jpg")
        cv2.imwrite(frame_filename, frame)
        print(f"Frame {frame_count} saved.")

    # Display the captured frame (optional, you can remove this if you don't need to see the video stream)
    cv2.imshow('Video Stream', frame)

    # Exit the loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close the window
cap.release()
cv2.destroyAllWindows()
```
